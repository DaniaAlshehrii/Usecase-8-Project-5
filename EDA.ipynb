```python
df.rename(columns={'الشركة': 'Company', 'الإفتتاح': 'Opening', 'الأعلى': 'Top', 'الأدنى': 'Lowest', 'الإغلاق': 'Closing'
                  , 'التغير': 'Change', '% التغيير': 'Change %', 'الكمية المتداولة': 'Traded Quantity', 
                   'القيمة المتداولة (ر.س)': 'Trading value (SAR)'}, inplace=True)
```

```python
pd.options.display.float_format = '{:,.2f}'.format

for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.replace(',', '').astype(float, errors='ignore')
```

```python
df.head()
```

```python
numerical_features = ['Opening', 'Top', 'Lowest', 'Closing', 'Change', 'Change %', 'Traded Quantity', 'Trading value (SAR)']
df[numerical_features].hist(figsize=(16, 20), bins=50)
```

```python
top_10_df = df.sort_values(by='Change', ascending=False).head(10)
fig = px.bar(top_10_df, x='Company', y='Change',
             title='Top 10 Changes Between Opening and Closing Prices',
             labels={'Entity': 'Entity', 'Change': 'Change'}
             )

# Show the plot
fig.show()
```

```python
top_5_df = df.sort_values(by='Top', ascending=False).head(5)

# Create a pie chart
fig = px.pie(top_5_df, names='Company', values='Top',
             title='Top 5 Companies by Top Value Distribution',)
             

# Show the plot
fig.show()
```

```python
top_10_opening = df.sort_values(by='Opening', ascending=False).head(10)

# Create a bar chart
fig = px.bar(top_10_opening, x='Company', y='Opening',
             title='Top 10 Companies by Opening Value',
             labels={'Company': 'Company', 'Opening': 'Opening Value'}             
             )

# Show the plot
fig.show()
```

```python
lowest_10_opening = df.sort_values(by='Opening', ascending=False).tail(10)

# Create a bar chart
fig = px.bar(lowest_10_opening, x='Company', y='Opening',
             title='lowest 10 Companies by Opening Value',
             labels={'Company': 'Company', 'Opening': 'Opening Value'},
             )

# Show the plot
fig.show()
```

```python
df.to_csv('clean_data.csv', index=False)
```
